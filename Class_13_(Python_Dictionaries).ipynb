*Dictionary Views*

Dictionary methods items, keys and values each return a view of a dictionaryâ€™s data. When you iterate over a view, it â€œseesâ€ the dictionaryâ€™s current contentsâ€”it does not have its own copy of the data.

To show that views do not maintain their own copies of a dictionaryâ€™s data, letâ€™s first save the view returned by keys into the variable months_view, then iterate through it:

```
months_data = months.keys()
print(months_data)


for key in months_data:
  print(key, end=' ')
```

```
months_data = months.items()
print(months_data)


for key in months_data:
  print(key, end=' ')
```

Next, letâ€™s add a new keyâ€“value pair to months and display the updated dictionary:

```
months['December'] = 12

months
```

Now, letâ€™s iterate through months_view again. The key we added above is indeed displayed:

```
for key in months_view:
  print(key, end=' ')
```

Do not modify a dictionary while iterating through a view. According to the Python Standard Library documentation, either youâ€™ll get a RuntimeError or the loop might not process all of the viewâ€™s values.

**1.2.7  Dictionary Comprehensions**

Dictionary comprehensions provide a convenient notation for quickly generating dictionaries, often by mapping one dictionary to another.



For example, in a dictionary with unique values, you can reverse the keyâ€“value pairs:

```
months = {'January': 1, 'February': 2, 'March': 3}

```

```
months2 = {number: name for name, number in months.items()}

months2
```

Curly braces delimit a dictionary comprehension, and the expression to the left of the for clause specifies a keyâ€“value pair of the form key: value. The comprehension iterates through months.items(), unpacking each keyâ€“value pair tuple into the variables name and number. The expression number: name reverses the key and value, so the new dictionary maps the month numbers to the month names.

What if months contained duplicate values? 


```
months_of_the_year = {'January': '1', 'February': '1', 'March': '3','April':'3','may':'3'}
print(months_of_the_year)

months_of_the_year2={ number:name for name, number in months_of_the_year.items()}
#months_of_the_year2={ name:number for name, number in months_of_the_year.items()}

print(months_of_the_year2)

print(months_of_the_year)
```

A dictionary comprehension also can map a dictionaryâ€™s values to new values. The following comprehension converts a dictionary of names and lists of grades into a dictionary of names and grade-point averages. The variables k and v commonly mean key and value:



```
grades = {'Sue': [98, 87, 94], 'Bob': [84, 95, 91]}

grades2 = {k: sum(v) / len(v) for k, v in grades.items()}

grades2
```

The comprehension unpacks each tuple returned by grades.items() into k (the name) and v (the list of grades). Then, the comprehension creates a new keyâ€“value pair with the key k and the value of sum(v) / len(v), which averages the listâ€™s elements.

**1.2.8 Example: Dictionary of Student Grades**

The script represents an instructorâ€™s grade book as a dictionary that maps each studentâ€™s name (a string) to a list of integers containing that studentâ€™s grades on three exams. In each iteration of the loop that displays the data , we unpack a keyâ€“value pair into the variables name and grades containing one studentâ€™s name and the corresponding list of three grades. Uses built-in function sum to total a given studentâ€™s grades, then calculates and displays that studentâ€™s average by dividing total by the number of grades for that student (len(grades)). Keep track of the total of all four studentsâ€™ grades and the number of grades for all the students, respectively. Prints the class average of all the studentsâ€™ grades on all the exams.

```
"""Using a dictionary to represent an instructor's grade book."""

grade_book = {'Susan': [92, 85, 100],'Eduardo': [83, 95, 79],'Azizi': [91, 89, 82],'Pantipa': [97, 91, 92]}

all_grades_total = 0

all_grades_count = 0

for name, grades in grade_book.items():
  total = sum(grades)
  print(f'Average for {name} is {total/len(grades):.2f}')
  all_grades_total += total
  all_grades_count += len(grades)

print(f"Class's average is: {all_grades_total / all_grades_count:.2f}")
```

**1.2.9 Example: Word Counts**

The script builds a dictionary to count the number of occurrences of each word in a string. Create a string text that weâ€™ll break into wordsâ€”a process known as tokenizing a string. Python automatically concatenates strings separated by whitespace in parentheses.Creates an empty dictionary. The dictionaryâ€™s keys will be the unique words, and its values will be integer counts of how many times each word appears in text.

```
"""Tokenizing a string and counting unique words."""

text = ('this is sample text with several words this is more sample text with some different words')

word_counts = {}

# count occurrences of each unique word
for word in text.split():
  
  if word in word_counts:
    # update existing key-value
    word_counts[word] += 1
  else:
    # insert new key-value pair
    word_counts[word] = 1

print(f'{"WORD":<12}COUNT')

for word, count in sorted(word_counts.items()):
  print(f'{word:<12}{count}')

print('\nNumber of unique words:', len(word_counts))
```

Tokenizes text by calling string method split, which separates the words using the methodâ€™s delimiter string argument. If you do not provide an argument, split uses a space. The method returns a list of tokens (that is, the words in text). Iterate through the list of words. For each word,determines whether that word (the key) is already in the dictionary. If so, increments that wordâ€™s count; otherwise, inserts a new keyâ€“value pair for that word with an initial count of 1.

Summarize the results in a two-column table containing each word and its corresponding count. The for statement iterates through the dictionaryâ€™s keyâ€“value pairs. It unpacks each key and value into the variables word and count, then displays them in two columns. Displays the number of unique words.

**1.2.10 Example: Word Counts (Python Library Module collections)**

The Python Standard Library already contains the counting functionality that we implemented using the dictionary and the loop. The module collections contains the type Counter, which receives an iterable and summarizes its elements. Letâ€™s reimplement the preceding script in fewer lines of code with Counter:

```
import collections

text = ('this is sample text with several words this is more sample text with some different words')

counter = collections.Counter(text.split())
print(counter)


for word, count in sorted(counter.items()):
  print(f'{word:<12}{count}')

print('Number of unique keys:', len(counter.keys()))
```

Creates the Counter, which summarizes the list of strings returned by text.split(). Counter method items returns each string and its associated count as a tuple. Use built-in function sorted to get a list of these tuples in ascending order. By default sorted orders the tuples by their first elements. If those are identical, then it looks at the second element, and so on. The for statement iterates over the resulting sorted list, displaying each word and count in two columns.
