# One hot encoding for some columns

Here are various methods for concatenating one-hot encoded columns with other numeric columns.

I have reused the month and hour columns from the bike dataset for ease of reuse.


```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'col1' : np.random.random(24),
    'mnth' : np.repeat(np.arange(12), 2),
    'hr'  : np.arange(24)
})
df
```

```python
np.repeat(np.arange(12), 2)
```

## One hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

enc = OneHotEncoder()

mnth_enc = enc.fit_transform(df[['mnth']])
hr_enc = enc.fit_transform(df[['hr']])
```

```python
enc.fit_transform(df)
```

First attempt at merging. Use the merge methods at csr_matrix.

```python
type(mnth_enc)
```

```python
from scipy.sparse import hstack
hstack([mnth_enc, hr_enc])
```

## Second method: convert to dataframe

```python

mnth_df = pd.DataFrame.sparse.from_spmatrix(mnth_enc)
hr_df = pd.DataFrame.sparse.from_spmatrix(hr_enc)
mnth_df.shape, hr_df.shape
```

```python
pd.concat([df.col1, mnth_df, hr_df], axis = 1)
```

## Third way: use column transformer utility from scikit

```python
enc.fit_transform(df[['mnth', 'hr']])
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()


ct = make_column_transformer(
    (ohe, ['mnth', 'hr' ]),
    ('passthrough', ['col1']) )
X = ct.fit_transform(df)
# notice that col1 is moved to the end because of the order in the column transformer.
pd.DataFrame.sparse.from_spmatrix(X)
```
